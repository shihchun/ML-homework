## 3.1

多元線性回歸與單元線性回歸（ex： least square最小平方回歸）類似，如果需要做的是比較不同自變量（屬性值）對因變量的影響強弱，也不需要考慮常數項，這樣得到的回歸係數是標準化的回歸係數，方法也可減去第一個樣本。

其他分類問題一般還是需要考慮偏置項的，$b$如果代表$y$的截距，沒有考慮一定是不行的。

機器學習學習出$w$和$b$，一般使用向量（vector is function）形式寫成入下式：
$$
f(\boldsymbol{x})=\boldsymbol{w}^{\mathrm{T}} \boldsymbol{x}+b
$$



$y=e^{w^{\mathrm{T}} x+b}$（對數線性）、
$y=\boldsymbol{w}^{\mathrm{T}} \boldsymbol{x}+b$（線性）


![From www.saedsayad.com](https://www.saedsayad.com/images/LogReg_1.png)

在對數線性回歸裏面$b_0$對於線是左右移動的，而$b_1$則是邊坡坡度（$b_0$是偏置項bias，其他是$w_n$或$w^*$（weight））

$$y=e^{w^{\mathrm{T}} x+b}（對數線性）$$

對兩邊取自然對數得到：

$$\ln y=\boldsymbol{w}^{\mathrm{T}} \boldsymbol{x}+b$$

令$y$爲勝算比（odds ratio），$y= \frac{p}{1-p}$、 $p$:為成功的機率，$(1-p)$:為失敗的機率，

$$
p=\frac{1}{1+e^{-\left(b_{0}+b_{1} x_{1}+b_{2} x_{2}+\cdots+b_{p} x_{p}\right)}} 
$$

## 3.3 Logistic regression

[sklean auto_examples](https://scikit-learn.org/stable/auto_examples/linear_model/plot_iris_logistic.html#sphx-glr-auto-examples-linear-model-plot-iris-logistic-py)

````
#@title Default title text
# Logistic Regression

'''
data importion
'''
import numpy as np  # for matrix calculation
import matplotlib.pyplot as plt  
import matplotlib

# 在windows上設定中文字型
# matplotlib.rc('font', family='mingliu', weight='bold') # 設定新細明體
# matplotlib.rc('font', family='simsun', weight='bold') # 設定宋體

# load the CSV file as a numpy matrix

dataset = np.loadtxt('/content/drive/My Drive/Colab Notebooks/watermelon_3a.csv', delimiter=",")
# separate the data from the target attributes
X = dataset[:,1:3] # 除去編號，找第2到第3個col
y = dataset[:,3] # 西瓜好壞 binary coding
np.shape(dataset) # (17,4)
m,n = np.shape(X) # (17,2) 

# draw scatter diagram to show the raw data
f1 = plt.figure(1)       
plt.title('watermelon_3a ori')  
plt.xlabel('density')  
plt.ylabel('ratio_sugar')  
plt.scatter(X[y == 0,0], X[y == 0,1], marker = 'o', color = 'k', s=100, label = 'bad')
plt.scatter(X[y == 1,0], X[y == 1,1], marker = 'o', color = 'g', s=100, label = 'good')
plt.legend(loc = 'upper right')
plt.show()

''' 
using sklearn lib for logistic regression
'''
from sklearn import metrics
from sklearn import model_selection
from sklearn.linear_model import LogisticRegression
import matplotlib.pylab as pl

# generalization of test and train set
# your code here
# 不知道為什麼test_size 最多只能帶入14個sample，sklearn預設範例數值是0.5，17個sample去學習出一半的數組
X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, test_size=14, random_state=0)

# model validation  ---> instantiated the function model
log_model = LogisticRegression(C=1e5, solver='lbfgs', multi_class='multinomial') 
log_model.fit(X_train, y_train) 



# model training
'''
model.fit(X,y,sample_weight)
  X: {arrary-like, sparse matrix}, shape(n,n)
  y: array_weight: array-like, shape(n_samples,)
  sample_weight: is optional using unit weight in default
'''
log_model.fit(X_train, y_train)


# summarize the fit of the model
y_pred = log_model.predict(X_test) # -->use Z to plot

# # 畫線測試
# y_sp = np.zeros((17,2))
# #np.disp(log_model.coef_) # [[0.15872911 0.54971927]]
# Cdict = { i : log_model.coef_[0,i] for i in range(0, len(log_model.coef_[0,:]) ) }
# print("coef_ 係數： ",Cdict)
# for i in range(17):
#   y_sp[i,0] = X[i,1]
#   y_sp[i,1] = Cdict[0]*X[i,1]+Cdict[1]*X[i,1]
#   pass
# #print("y_sp: \n",y_sp,"\n",y, "\n X: \n", X)
# plt.plot(y_sp[:,0], y_sp[:,1]) #對應每個密度的西瓜找到一個回歸後的數值

# Put the result into a color plot

# Plot also the training pointsplt.title('watermelon_3a')  


f2 = plt.figure(2)       
plt.title('watermelon_3a')  
plt.xlabel('density')  
plt.ylabel('ratio_sugar')

x_min, x_max = X[:, 0].min() - .5, X[:, 0].max() + .5
y_min, y_max = X[:, 1].min() - .5, X[:, 1].max() + .5
h = 0.002  # step size in the mesh
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
Z = log_model.predict(np.c_[xx.ravel(), yy.ravel()])

# Put the result into a color plot
Z = Z.reshape(xx.shape)
plt.pcolormesh(xx, yy, Z, cmap=plt.cm.Paired)

# Plot also the training points
plt.scatter(X[y == 0,0], X[y == 0,1], marker = 'o', color = 'k', s=100, label = 'bad')
plt.scatter(X[y == 1,0], X[y == 1,1], marker = 'o', color = 'g', s=100, label = 'good')
plt.legend(loc = 'upper right')


plt.show()

# plt.show()

print("confusion_matrix:\n " , metrics.confusion_matrix(y_test, y_pred))
print("classification:\n" , metrics.classification_report(y_test, y_pred))
print("y: ",np.shape(y) ,"\n", y, "\ny_pred: ",np.shape(y_pred) ,"\n", y_pred)
print("X: ",np.shape(X) ,"\n", X, "\nX_test: ",np.shape(X_test) ,"\n", X_test)
print(np.shape(X_test),np.shape(X_train),np.shape(y_pred))
````

## 3.5 LDA

````
'''
LDA via sklearn
'''


import matplotlib.pyplot as plt

# load the CSV file as a numpy matrix

dataset = np.loadtxt('/content/drive/My Drive/Colab Notebooks/watermelon_3a.csv', delimiter=",")
# separate the data from the target attributes
X = dataset[:,1:3] # 除去編號，找第2到第3個col
y = dataset[:,3] # 西瓜好壞 binary coding
np.shape(dataset) # (17,4)
m,n = np.shape(X) # (17,2) 

# draw scatter diagram to show the raw data
f1 = plt.figure(1)       
plt.title('watermelon_3a ori')  
plt.xlabel('density')  
plt.ylabel('ratio_sugar')  
plt.scatter(X[y == 0,0], X[y == 0,1], marker = 'o', color = 'k', s=100, label = 'bad')
plt.scatter(X[y == 1,0], X[y == 1,1], marker = 'o', color = 'g', s=100, label = 'good')
plt.legend(loc = 'upper right')
plt.show()

from sklearn import model_selection
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn import metrics

# generalization of train and test set
# 不知道為什麼test_size 最多只能帶入14個sample，sklearn預設範例數值是0.5，17個sample去學習出一半的數組
X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, test_size=14, random_state=0)
# model fitting

lda_model = LinearDiscriminantAnalysis() 
lda_model.fit(X_train, y_train)

# model validation

# summarize the fit of the model
y_pred = lda_model.predict(X_test) # -->use Z to plot



# Put the result into a color plot

# Plot also the training pointsplt.title('watermelon_3a')  

f2 = plt.figure(2)
plt.title('watermelon_3a')
plt.xlabel('density')
plt.ylabel('ratio_sugar')

x_min, x_max = X[:, 0].min() - .5, X[:, 0].max() + .5
y_min, y_max = X[:, 1].min() - .5, X[:, 1].max() + .5
h = 0.002  # step size in the mesh
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
Z = lda_model.predict(np.c_[xx.ravel(), yy.ravel()])

# Put the result into a color plot
Z = Z.reshape(xx.shape)
plt.pcolormesh(xx, yy, Z, cmap=plt.cm.Paired)

# Plot also the training points
plt.scatter(X[y == 0,0], X[y == 0,1], marker = 'o', color = 'k', s=100, label = 'bad')
plt.scatter(X[y == 1,0], X[y == 1,1], marker = 'o', color = 'g', s=100, label = 'good')
plt.legend(loc = 'upper right')


plt.show()

print("confusion_matrix:\n " , metrics.confusion_matrix(y_test, y_pred))
print("classification:\n" , metrics.classification_report(y_test, y_pred))
print("y: ",np.shape(y) ,"\n", y, "\ny_pred: ",np.shape(y_pred) ,"\n", y_pred)
print("X: ",np.shape(X) ,"\n", X, "\nX_test: ",np.shape(X_test) ,"\n", X_test)
print(np.shape(X_test),np.shape(X_train),np.shape(y_pred))
````
